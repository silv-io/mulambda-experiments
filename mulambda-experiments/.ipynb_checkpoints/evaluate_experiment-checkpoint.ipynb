```python
ev = gw.events(exp_id)
ev.head(10)
```

```python
import json
from util import json_transform


copy = ev.copy()
copy['value'] = copy['value'].apply(json.loads)
transformed = json_transform(copy)
transformed
```

```python
transformed.plot(x='ts', y='elapsed')
```

```python
tm[tm["metric"] == "cpu"].plot(x='ts', y='value')
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, AutoDateLocator

df = tm[tm["metric"] == "cpu"]
plt.figure(figsize=(10, 6))
plt.plot(df["ts"], df["value"])

# Format the x-axis timestamp display
date_format = DateFormatter("%b %d %H:%M")
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
locator = AutoDateLocator(maxticks=5)
plt.gca().xaxis.set_major_locator(locator)

plt.xlabel("Time")
plt.ylabel("CPU")
plt.grid(True)
plt.show()
```
